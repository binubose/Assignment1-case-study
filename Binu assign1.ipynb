```python
plt.figure(figsize=(10,10))
plt.subplot(3,1,1)
plt.bar(data['Classification'],data['SW'])
plt.title('Classification VS SW',fontsize=20)
plt.xlabel('Classification',fontsize=15)
plt.ylabel('SW')
plt.xticks(rotation=90)

plt.figure(figsize=(10,10))
plt.subplot(3,1,2)
plt.bar(data['Classification'],data['PL'])
plt.title('Classification VS SW',fontsize=20)
plt.xlabel('Classification',fontsize=15)
plt.ylabel('PL')
plt.xticks(rotation=90)

plt.figure(figsize=(10,10))
plt.subplot(3,1,3)
plt.bar(data['Classification'],data['SL'])
plt.title('Classification VS SW',fontsize=20)
plt.xlabel('Classification',fontsize=15)
plt.ylabel('SL')
plt.xticks(rotation=90)
          
plt.tight_layout()
```
